```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

# Function to scrape details from a single book's page
def get_book_details(book_url):
    """Scrapes various details of a book from its Goodreads page.

    Args:
        book_url (str): The URL of the Goodreads book page.

    Returns:
        dict: A dictionary containing scraped information.
    """
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36"
    }
    response = requests.get(book_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract elements
    genre = soup.find('div', {'class': 'BookPageMetadataSection__genres'})
    date = soup.find('p', {'data-testid': 'publicationInfo'})
    author = soup.find('span', {'class': 'ContributorLink__name'})
    title = soup.find('h1', {'class': 'Text Text__title1'})
    rating = soup.find('div', {'class': 'RatingStatistics__column'})
    ratings = soup.find('span', {'data-testid': 'ratingsCount'})
    reviews = soup.find('span', {'data-testid': 'reviewsCount'})
    au_det = soup.find('div', {'class': 'FeaturedPerson__infoPrimary'})
    # Clean up extracted data
    genre = genre.text.strip(',') if genre else 'Not available'
    date = date.text.strip() if date else 'Not available'
    title = title.text.strip() if title else 'Not available'
    rating = rating.text.strip() if rating else 'Not available'
    ratings = ratings.text.strip() if ratings else 'Not available'
    reviews = reviews.text.strip() if reviews else 'Not available'
    author = author.text.strip() if author else 'Not available'
    au_det = au_det.text.strip() if au_det else 'Not available'

    # Return data as a dictionary
    return {
        "Genre": genre,
        "Publication Date": date,
        "Author": author,
        "Title": title,
        "Rating": rating,
        "Ratings Count": ratings,
        "Reviews Count": reviews,
        "Author Info": au_det

    }

# Base URL for the book list
base_url = 'https://www.goodreads.com/list/show/6.Best_Books_of_the_20th_Century?page='
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36"
}

book_urls = []
count=0
# Loop through pages to collect book URLs
for page_num in range(60, 80):  # Loop through pages 1 to 80
    url = base_url + str(page_num)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    # No need to loop through <tr> tags, directly find book titles
    urls = soup.find_all('a', class_='bookTitle')  # Get all book URLs on the page
    for url in urls:
        book_url = "https://www.goodreads.com" + url['href']  # Extract href for each URL
        book_urls.append(book_url)  # Append to the list
# List to hold all book data

all_books = []

# Loop through each book URL to collect data
for book_url in book_urls:
    print(f"Scraping data from {book_url}...")
    count=count+1
    print(count)
    details = get_book_details(book_url)

    # Store the scraped data
    all_books.append(details)
```

```python
if all_books:
    df = pd.DataFrame(all_books)
    df.to_csv('goodreads_books2.csv', index=False)
    print("Scraping complete. Data saved to goodreads_books2.csv.")
else:
    print("No data was collected.")
```
